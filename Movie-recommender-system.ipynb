```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credit = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credit.head()
```

# Merging both dataset on the basis of 'titles' column

```python
movies = movies.merge(credit , on='title')
movies.head(1)
```

# Keeping only these columns , rest are not important

```python
movies = movies[['genres', 'id', 'keywords', 'title', 'overview', 'cast', 'crew']]
```

```python
movies.info()
```

# Checking the null values

```python
movies.isnull().sum()
```

# Dropping the null values

```python
movies.dropna(inplace = True)
```

# Checking the duplicate values

```python
movies.duplicated().sum()
```

# Putting the data in right form as we need keywords

```python
movies.loc[0].genres
```

```python
import ast
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies
```

```python
movies.loc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies
```

```python
movies.loc[0].cast
```

```python
def convert3(obj):
    counter = 0
    l = []
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i['name'])
            counter += 1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies
```

# Removing spaces

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" " , "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" " , "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" " , "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" " , "") for i in x])
```

```python
movies
```

# Merging everything in tag columns and removing other columns

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['title' , 'id' , 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : " ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
x = cv.fit_transform(new_df['tags']).toarray()
```

```python
x
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem('loved')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df
```

# Now will find distance but will not use euclidean distnace , rather we will use cosine distance

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(x)
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)) , reverse = True , key = lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
